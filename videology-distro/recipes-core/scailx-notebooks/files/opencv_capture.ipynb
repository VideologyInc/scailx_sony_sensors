```python
import cv2
import numpy as np
from glob import glob
from PIL import Image
from IPython.display import display, clear_output
```

```python
cams = glob("/dev/video-i*")
cam_dev = cams[0]
print(f"using cam device {cam_dev}") 
res = (1280, 720)
cam = cv2.VideoCapture(f'v4l2src device={cam_dev} ! video/x-raw,width={res[0]},height={res[1]} ! imxvideoconvert_g2d ! video/x-raw,format=(string)RGBA ! appsink', cv2.CAP_GSTREAMER)
ret, img = cam.read()
if img is None:
    print("something went wrong")
else: 
    display(Image.fromarray(img))
cam.release()
```
